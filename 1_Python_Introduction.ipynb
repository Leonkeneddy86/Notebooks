```python
# Condicionales
lenguaje = "Python"

if lenguaje == "Python":
    print("Es un lenguaje interpretado")
elif lenguaje == "Go" or lenguaje == "C":
    print("Es un lenguaje compilado")
else:
    print("¿Es Java?")

```

### Bucles

#### Bucles for

```python
"""
Funciona para iterar cadenas de texto, listas
tuplas, diccionarios y sets
"""
mensaje = "Hola Promocion 5 de IA"
print(len(mensaje))
for interator in mensaje:
    print(interator)
```

```python
"""
En combinación con la función range()
para iterar con un número como límite
"""

for i in range(1,7):
    print(i)
```

```python
"""
Cuando se usa la función range()
se puede indicar el punto de inicio, final y el paso
"""
for i in range(5, 25, 3):
    print(i)
```

```python
"""
For else, nos permite agregar una condición sí no,
cuando se puede romper la iteración con break
"""
for i in [1, 2, 3, 4, 5, 6]:
    if i == 5:
        print("El conteo terminó")
        print(i)
        break

    else:
        print("El conteo continua")
        print(i)
```

#### Bucle while


```python
respuesta = False

while respuesta == False:
    nombre = input("¿Cuál es tu nombre? ")
    if nombre.lower() == "luis":
        print("Nombre correcto")
        respuesta = True
    else:
        print("Nombre equivocado")

```

# Calculadora

```python
def suma(a, b):
  return a + b

def resta(a, b):
  return a - b

def multiplicacion(a, b):
  return a * b

def division(a, b):
  if b != 0:
    return a / b
  else:
    return "Error: División por cero"
```

```python
def calculadora():
  print("Bienvenido a la calculadora")
  print("Operaciones disponibles")
  print("1. Suma")
  print("2. Resta")
  print("3. Multiplicación")
  print("4. División")

  while True:
    opcion = __builtins__.input("Seleccione una operación (1-4) o 'q' para salir: ")

    if opcion == 'q':
      print("Gracias por usar la calculadora, ¡Nos vemos pronto!")
      break

    if opcion not in ['1', '2', '3', '4']:
      print("Opción no válida. Por favor, intente de nuevo.")
      continue

    num_1 = float(__builtins__.input("Ingrese un número: "))
    num_2 = float(__builtins__.input("Ingrese otro número: "))

    if opcion == '1':
      print(f"Resultado: {suma(num_1, num_2)}")
    elif opcion == '2':
      print(f"Resultado: {resta(num_1, num_2)}")
    elif opcion == '3':
      print(f"Resultado: {multiplicacion(num_1, num_2)}")
    elif opcion == '4':
      print(f"Resultado: {division(num_1, num_2)}")

if __name__== '__main__':
   calculadora()
```
